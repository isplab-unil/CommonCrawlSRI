```python
import csv
from collections import defaultdict

total = 0
groups = defaultdict(int)

with open('../notes/top-urls.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        try: 
            
            number = int(row['number'])
            total += number           
            library = row['library']

            if "bootstrap" in library:
                groups[('library', 'bootstrap')] += number
            
            elif "jquery" in library:
                groups[('library', 'jquery')] += number
                
            elif "fontawesome" in library or "font-awesome" in library:
                groups[('library', 'fontawesome')] += number
                
            elif library.startswith("cdnjs.cloudflare.com"):
                group = library.split('/')[3]
                groups[('library', group)] += number
                
            elif library.startswith("ajax.googleapis.com"):
                group = library.split('/')[3]
                groups[('library', group)] += number
                
            elif library.startswith("unpkg.com"):
                group = library.split('/')[1].split('@')[0]
                groups[('library', group)] += number
                
            elif library.startswith("cdn.jsdelivr.net/npm"):
                group = library.split('/')[2].split('@')[0]
                groups[('library', group)] += number
                
            elif library.startswith("cdn.jsdelivr.net/gh"):
                group = library.split('/')[3].split('@')[0]
                groups[('library', group)] += number
                
            elif library.startswith("cdn.jsdelivr.net/wp") or library.startswith("cdn.jsdelivr.net/wordpress"):
                groups[('library', 'wordpress')] += number
            
            #elif library.startswith("cdn.jsdelivr.net"):
                #group = library.split('/')[1]
                #groups[('library', group)] += number
            
            #else:
                group = library.split('/')[0]
                #if not "." in group:
                    #groups[('domain', 'local')] += number
                #else:
                    #groups[('domain', group)] += number
        except:
            print(row)
        

```
