# 08 Top-k SRI files

```python
import re
from glob import iglob
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.shell import sqlContext
```

## Load the CSV files

```python
data = sqlContext.read.parquet('../data/sri/2019-09/reports/08_topk_sri_url/').toPandas()
```

## Display the data

## Plot the data

```python
fig, ax = plt.subplots(1, 1)

data['percentage'].plot.bar(ax=ax, sharex=True, rot='xticks')
#data['percentage'].cumsum().plot.line(ax=ax, sharex=True)

plt.xticks(rotation=90)
plt.xlim(-0.5,9.5)
plt.xlabel('')
plt.ylabel('SRI')

ax.set_yticklabels(['{:,.0%}'.format(x / 100) for x in ax.get_yticks()])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid()

plt.savefig('08_topk_sri_domain.pdf', bbox_inches='tight')
```

```python
sqlContext.read.parquet('../data/sri/2019-09/reports/08_topk_sri_url/') \
    .repartition(1) \
    .write.format("com.databricks.spark.csv") \
    .option("quoteAll", "true") \
    .option("header", "true") \
    .save("08_topk_sri_url.csv")
```
